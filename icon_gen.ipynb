```python
from PIL import Image
import numpy as np

def shrink_gradient_pad_with_offset(image_path, 
                                     output_path, 
                                     scale=0.5, 
                                     horizontal_offset=0.0, 
                                     vertical_offset=0.0):
    """
    缩小图片，用原图上下边界颜色生成垂直渐变背景，并生成正方形。
    可以指定缩小后图片在画布上的水平和垂直偏移，允许图片移出画布。
    
    参数:
        image_path (str): 输入图片路径
        output_path (str): 输出图片路径
        scale (float): 缩小比例 (0~1之间)
        horizontal_offset (float): 水平偏移量。负值向左偏移，正值向右偏移。
        vertical_offset (float): 垂直偏移量。负值向上偏移，正值向下偏移。
    """
    # 1. 打开图像并获取基本信息
    try:
        img = Image.open(image_path).convert("RGB")
    except FileNotFoundError:
        print(f"错误: 文件 '{image_path}' 未找到。")
        return
        
    w, h = img.size
    
    # 2. 提取原图上下边界的平均颜色
    np_img = np.array(img)
    top_avg_color = np.mean(np_img[0, :, :], axis=0)
    bottom_avg_color = np.mean(np_img[-1, :, :], axis=0)
    
    # 3. 计算缩小后尺寸和画布尺寸
    new_w, new_h = int(w * scale), int(h * scale)
    max_side = max(w, h)
    
    # 4. 根据偏移参数计算粘贴位置
    # 允许图片完全移出画布，偏移范围不再限制
    # 计算偏移量：偏移值乘以图片尺寸，使偏移效果更明显
    x_offset = int(max_side/2 - new_w/2 + horizontal_offset * max_side)
    y_offset = int(max_side/2 - new_h/2 + vertical_offset * max_side)
    
    # 5. 创建渐变背景
    gradient_canvas = np.zeros((max_side, max_side, 3), dtype=np.uint8)
    
    # 生成整个画布的渐变背景，不再局限于图片位置
    for y in range(max_side):
        t = y / (max_side - 1) if max_side > 1 else 0.5
        color = (1 - t) * top_avg_color + t * bottom_avg_color
        gradient_canvas[y, :, :] = color.astype(np.uint8)
    
    # 将 NumPy 数组转回 PIL Image
    canvas = Image.fromarray(gradient_canvas)
    
    # 6. 将缩小后的图片粘贴到画布的指定位置
    img_resized = img.resize((new_w, new_h), Image.Resampling.LANCZOS)
    
    # 直接粘贴，允许图片部分或完全移出画布
    canvas.paste(img_resized, (x_offset, y_offset))
    
    # 7. 保存结果
    canvas.save(output_path)
    print(f"图片已处理并保存到 {output_path}")


if __name__ == "__main__":
    try:
        icon = 'icon-1h9l2pa65scf4d92g1xzs0x7t8i.png'
        # 使用更大的偏移值使效果更明显
        shrink_gradient_pad_with_offset(f"ori_icon/{icon}", f"web_icon/{icon}", 
                                       scale=1, vertical_offset=-0.035)  # 尝试 -0.5 到 -1.5 的值
    except FileNotFoundError:
        print("错误: f'{icon}' 未找到。请确保图片文件存在于脚本运行的目录中。")
```
